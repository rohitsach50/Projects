```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text = requests.get('https://www.imdb.com/search/title/?groups=oscar_winner,best_picture_winner&sort=year,desc&count=250').content
soup = BeautifulSoup(html_text, 'lxml')
```

```python
details = q.p.text.replace('\n', '')
details
```

```python
certificate_str = ""
pttrn = re.compile(r'[A-ZA-Z\-\d\d]{2}\|')
certificate = pttrn.findall(details)
for i in certificate:
    certificate_str+=i.replace("|","")
    
certificate_str
```

```python
g = q.p.find('span', class_='certificate')
g.text
```

```python
runtime_str=""
pttrn = re.compile(r'\|.*\|')
runtime = pttrn.findall(details)
for i in runtime:
    runtime_str+=i.replace("|","")
runtime_str
```

```python
genre_str=""
pttrn = re.compile(r'\|[A-Za-z]*')
genre = pttrn.findall(details)
for i in genre:
    genre_str+=i.replace("|","")
genre_str
```
