```python
import re
```

```python
txt = open("test.txt", "r")
a = txt.read()
txt.close
print(a)
```

```python
years_list = []
pattern = re.compile(r'year..\d\d\d\d.')
matches = pattern.findall(a)
for match in matches:
      
    years_list.append(match)  
    print(match)
```

```python

years_list
```

```python
yrs_lst=""
for i in years_list:
    yrs_lst += i
```

```python
yrs_lst
```

```python
y = []
pttrn = re.compile(r'\d\d\d\d')
yrs = pttrn.findall(yrs_lst)
for yr in yrs:
    y.append(yr)
y

```

```python
yl=[]
for i in y:
    yl.append(int(i))
yl

```
